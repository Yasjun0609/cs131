```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Yasjun0609/cs131/master/MiniProjectGroup4/Partially%20Cleaned%20Salary%20Dataset.csv")
```

```python
data2 = pd.read_csv("https://raw.githubusercontent.com/Yasjun0609/cs131/master/MiniProjectGroup4/PartiallyCleanedSalaryDatasetNoOutlier.csv")
```

Job Title vs Salary Relplot

```python
chart = sns.relplot(data=data, x="Job Title", y="Salary", hue="Location", size="Salaries Reported")
plt.ylabel('Salary per Millions')
plt.title("Company vs Salary by Job Title", pad=1.05);
plt.xticks(rotation=-90, ha='center', fontsize=10)
```

Average Salary per Location Bar Graph

Average Salary per Job Title Reported Bar Graph

```python
selected_columns = data[['Job Title', 'Salary']]

average_salary_by_title = selected_columns.groupby('Job Title').mean()

average_salary_by_title = average_salary_by_title.sort_values(by='Salary', ascending=True)

average_salary_by_title.plot(kind='bar', legend=False)
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.xticks(rotation=45, ha='right')  # Rotating x-axis labels for better readability
plt.show()
```

```python
selected_columns = data2[['Job Title', 'Salary']]

average_salary_by_title = selected_columns.groupby('Job Title').mean()

average_salary_by_title = average_salary_by_title.sort_values(by='Salary', ascending=True)

average_salary_by_title.plot(kind='bar', legend=False)
plt.title('Average Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.xticks(rotation=45, ha='right')  # Rotating x-axis labels for better readability
plt.show()
```
